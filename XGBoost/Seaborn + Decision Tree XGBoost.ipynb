### About the Dataset:

The [Breast Cancer Diagnostic data](https://archive.ics.uci.edu/ml/datasets/Breast+Cancer+Wisconsin+%28Diagnostic%29) is available on the UCI Machine Learning Repository. 
Features are computed from a digitized image of a fine needle aspirate (FNA) of a breast mass. They describe characteristics of the cell nuclei present in the image. n the 3-dimensional space is that described in: [K. P. Bennett and O. L. Mangasarian: "Robust Linear Programming Discrimination of Two Linearly Inseparable Sets", Optimization Methods and Software 1, 1992, 23-34].

**Attribute Information**:

- ID number
- Diagnosis (M = malignant, B = benign)

Ten real-valued features are computed for each cell nucleus:

1. radius (mean of distances from center to points on the perimeter) 
2. texture (standard deviation of gray-scale values) 
3. perimeter 
4. area 
5. smoothness (local variation in radius lengths) 
6. compactness (perimeter^2 / area - 1.0) 
7. concavity (severity of concave portions of the contour) 
8. concave points (number of concave portions of the contour)
9. symmetry
10. fractal dimension ("coastline approximation" - 1)

The mean, standard error and "worst" or largest (mean of the three largest values) of these features were computed for each image, resulting in 30 features. For instance, field 3 is Mean Radius, field 13 is Radius SE, field 23 is Worst Radius.

All feature values are recoded with four significant digits.

Missing attribute values: none

Class distribution: 357 benign, 212 malignant

### Classification using XGBoost (minimal feature selection)

```python
from sklearn.model_selection import train_test_split
import xgboost as xgb
from sklearn.metrics import f1_score,confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
x_train, x_test, y_train, y_test = train_test_split(df, y, train_size = 0.7, random_state = 42)

clf_1 = xgb.XGBClassifier(random_state=42)
clf_1 = clf_1.fit(x_train,y_train)
```

```python
accuracy_score(y_test, clf_1.predict(x_test))
```

```python
cm = confusion_matrix(y_test, clf_1.predict(x_test))
sns.heatmap(cm, annot = True, fmt = 'd')
```

### Univariate Feature Selection and XGBoost

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
select_feature = SelectKBest(chi2, k = 13).fit(x_train, y_train)
```

```python
select_feature.scores_
```

```python
x_tr_2 = select_feature.transform(x_train)
x_ts_2 =  select_feature.transform(x_test)

clf_2 = xgb.XGBClassifier().fit(x_tr_2, y_train)
```

```python
accuracy_score(y_test, clf_2.predict(x_ts_2))
```

```python
cm = confusion_matrix(y_test, clf_2.predict(x_ts_2))
sns.heatmap(cm, annot = True, fmt = 'd')
```

### Recursive Feature Elimination with Cross-Validation

```python
from sklearn.feature_selection import RFECV

clf_3 = xgb.XGBRFClassifier()
rfecv = RFECV(estimator=clf_3, step = 1, cv = 10, scoring = 'accuracy').fit(x_train, y_train)
```

```python
rfecv.n_features_
```

```python
x_train.columns[rfecv.support_] # print best features
```

```python
accuracy_score(y_test, rfecv.predict(x_test) )
```

```python
cm = confusion_matrix(y_test, rfecv.predict(x_test))
sns.heatmap(cm, annot = True, fmt = 'd')
```

```python
num_features = [i for i in range(len(rfecv.grid_scores_))]
cv_scores = rfecv.grid_scores_
plt.figure(figsize=(10,10))
sns.lineplot(x = num_features, y = cv_scores, marker = 'X', color = 'r', markersize=16)
```

### Feature Extraction using Principal Component Analysis

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.28, random_state = 42)
x_train_norm = (x_train - x_train.mean()) / (x_train.max() - x_train.min())
x_test_norm = (x_test - x_test.mean()) / (x_test.max() - x_test.min())
```

```python
from sklearn.decomposition import PCA

pca = PCA()
pca.fit(x_train_norm)
```

```python
sns.lineplot(data = np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('# of components')
plt.ylabel('Explained variance')
```
